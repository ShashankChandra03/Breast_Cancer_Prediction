# Breast Cancer Detection

## Importing Libraries and Dataset

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

## Data Representation

```python
data = load_breast_cancer()
label_names = data["target_names"]
labels = data["target"]
feature_names = data["feature_names"]
features = data["data"]
```

```python
print(label_names)
print("Class label :", labels[0])
print(feature_names)
print(features[0], "\n")
```

## Splitting the Dataset

```python
train, test, train_labels, test_labels = train_test_split(features, labels,test_size=0.2, random_state=42)
```

## Using Logistic Regression for Breast Cancer Detection 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

train, test, train_labels, test_labels = train_test_split(features, labels, test_size=0.2, random_state=42)

logistic_model = LogisticRegression(max_iter=10000, random_state=42)
logistic_model.fit(train, train_labels)

predictions = logistic_model.predict(test)
```

## Calculating Accuracy of Logistic Regression

```python
accuracy = accuracy_score(test_labels, predictions)

print("Accuracy of Logistic Regression:", accuracy * 100, "%")
```

## Using Naive Bayes for Breast Cancer Detection

```python
gnb = GaussianNB()
gnb.fit(train, train_labels)
```

```python
preds = gnb.predict(test)
print(preds, "\n")
```

## Calculating Accuracy of Naive Bayes 

```python
print(accuracy_score(test_labels, preds))
```

## Comparison of Both Models

```python
import matplotlib.pyplot as plt

models = ['Naive Bayes', 'Logistic Regression']
accuracies = [0.9736, accuracy]

plt.figure(figsize=(8, 5))
plt.bar(models, accuracies, color=['blue', 'green'])
plt.ylim(0.9, 1.0)
plt.title('Comparison of Model Accuracies', fontsize=14)
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)

for i, acc in enumerate(accuracies):
    plt.text(i, acc - 0.005, f"{acc*100:.2f}%", ha='center', fontsize=10, color='white', weight='bold')

plt.show()
```
